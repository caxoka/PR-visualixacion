```python
import pandas as pd
```

```python
df_authors = pd.read_csv("./dataset/top10k-spotify-artist-metadata.csv")
df_authors
```

```python

# Limpiar: quitar duplicados y filas completamente vacías
df_authors_clean = df_authors.drop_duplicates().dropna(how='all')
df_authors_clean['artist'] = (
    df_authors_clean['artist']
    .str.strip()      
    .str.lower()     
)

df_authors_clean
```

```python
df_songs_2025=pd.read_csv("./dataset/spotify_data clean.csv")
df_songs_2025
```

```python
df_songs_2025_clean = df_songs_2025.drop_duplicates().dropna(how='all')
df_songs_2025_clean['artist_name'] = (
    df_songs_2025_clean['artist_name']
    .str.strip()      
    .str.lower()     
)

df_songs_2025_clean
```

## merge

```python
df_merged = df_songs_2025_clean.merge(df_authors_clean, left_on = "artist_name", right_on= "artist", how='inner')
df_merged
```

```python
df_merged.to_excel("dataset final.xlsx")
```
